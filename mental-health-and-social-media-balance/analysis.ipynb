# Descrição dataset

O conjunto de dados “Mental Health and Social Media Balance” contém 500 registros (usuários) e 10 colunas, explorando a relação entre o uso de redes sociais e fatores de saúde mental e bem-estar.

Cada linha representa um usuário individual, com informações sobre seus hábitos digitais, estilo de vida e indicadores emocionais.

| Coluna                        | Descrição                                                       |
| ----------------------------- | --------------------------------------------------------------- |
| **User_ID**                   | Identificador único de cada participante.                       |
| **Age**                       | Idade do usuário (16 a 49 anos).                                |
| **Gender**                    | Gênero do usuário: `Male`, `Female` ou `Other`.                 |
| **Daily_Screen_Time(hrs)**    | Tempo médio de tela diário (em horas).                          |
| **Sleep_Quality(1-10)**       | Avaliação da qualidade do sono (1 = péssimo, 10 = excelente).   |
| **Stress_Level(1-10)**        | Nível de estresse (1 = sem estresse, 10 = muito estressado).    |
| **Days_Without_Social_Media** | Número de dias sem uso de redes sociais.                        |
| **Exercise_Frequency(week)**  | Quantas vezes o usuário se exercita por semana.                 |
| **Social_Media_Platform**     | Plataforma mais usada (Facebook, Instagram, TikTok etc.).       |
| **Happiness_Index(1-10)**     | Índice de felicidade percebida (1 = infeliz, 10 = muito feliz). |

# Identificação de categorias (value_counts() e unique())

```python
# @title Selecionar colunas categóricas, exceto 'User_ID'
categorical_cols = [col for col in df.select_dtypes(include='object').columns if col != 'User_ID']
print("Colunas categóricas consideradas:", list(categorical_cols))
```

```python
# @title Exibir valores únicos e contagens de frequência
for col in categorical_cols:
    print(f"\n=== {col} ===")
    print("Valores únicos:", df[col].unique())
    print("\nContagem de valores:")
    print(df[col].value_counts())
```

# Análises agrupadas (groupby() e sort_values())

```python
# @title Média do índice de felicidade por plataforma

happiness_by_platform = (
    df.groupby("Social_Media_Platform")["Happiness_Index(1-10)"]
    .mean()
    .sort_values(ascending=False)
)

print("\nMédia de felicidade por plataforma:")
happiness_by_platform
```

### 💬 Insight:
Observa-se que usuários do X (Twitter) e LinkedIn apresentam os maiores índices médios de felicidade, seguidos por usuários de TikTok, Facebook e YouTube. Já os usuários do Instagram registram o menor nível médio de felicidade entre as plataformas analisadas.

Esse resultado pode indicar que plataformas com interações mais objetivas ou foco profissional (como o LinkedIn) tendem a favorecer um ambiente emocionalmente mais positivo, enquanto redes mais voltadas à comparação social e exposição pessoal (como o Instagram) podem estar associadas a menor bem-estar subjetivo.

```python
# @title Extra
happiness_by_platform = (
    df.groupby("Social_Media_Platform")["Happiness_Index(1-10)"]
    .mean()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))
plt.bar(happiness_by_platform.index, happiness_by_platform.values, color='skyblue')
plt.title("Felicidade Média por Plataforma de Mídia Social")
plt.xlabel("Plataforma")
plt.ylabel("Índice Médio de Felicidade (1-10)")
plt.xticks(rotation=30)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# @title Nível médio de estresse por faixa de tempo de tela

stress_by_screen = (
    df.groupby(pd.cut(df["Daily_Screen_Time(hrs)"], bins=[0,3,5,7,10]))["Stress_Level(1-10)"]
    .mean()
)

print("\nNível médio de estresse por faixa de tempo de tela (em horas):")
stress_by_screen
```

### 💬 Insight:
 Há uma clara tendência de aumento no nível de estresse conforme o tempo de tela cresce.
 Usuários com até 3h de tela por dia têm estresse médio em torno de 5,8,
 enquanto aqueles acima de 7h chegam próximos de 7,4.
 Isso indica uma correlação positiva entre uso prolongado de redes sociais e maior estresse percebido.

```python
# @title Extra
stress_by_screen = (
    df.groupby(pd.cut(df["Daily_Screen_Time(hrs)"], bins=[0,3,5,7,10]))["Stress_Level(1-10)"]
    .mean()
)

plt.figure(figsize=(7,4))
plt.plot(stress_by_screen.index.astype(str), stress_by_screen.values, marker='o', color='orange')
plt.title("Relação entre Tempo de Tela e Estresse Médio")
plt.xlabel("Faixa de Tempo de Tela (horas/dia)")
plt.ylabel("Nível Médio de Estresse (1-10)")
plt.grid(True)
plt.show()
```

```python
# @title # Felicidade média conforme frequência de exercícios por semana
happiness_by_exercise = (
    df.groupby("Exercise_Frequency(week)")["Happiness_Index(1-10)"]
    .mean()
    .sort_values(ascending=False)
)

print("\nFelicidade média por frequência de exercício semanal:")
happiness_by_exercise
```

### 💬 Insight:
A prática de exercícios está fortemente associada à felicidade.
 Usuários que se exercitam de 4 a 7 vezes por semana registram médias de felicidade acima de 9,
 enquanto pessoas sedentárias (0 ou 1 vez por semana) ficam em torno de 7.
 Isso reforça o papel do exercício físico como fator de bem-estar emocional e mental.

```python
# @title Extra

plt.figure(figsize=(7,4))
plt.scatter(df["Exercise_Frequency(week)"].values, df["Happiness_Index(1-10)"].values, alpha=0.6, color='green')
plt.title("Relação entre Exercício Semanal e Felicidade")
plt.xlabel("Frequência de Exercício (vezes por semana)")
plt.ylabel("Índice de Felicidade (1-10)")
plt.grid(True)
plt.show()
```

### A dispersão mostra tendência positiva conforme a frequência de exercícios aumenta.
 Os índices de felicidade tendem a ser mais altos.
 Usuários sedentários (0-1x por semana) concentram-se em níveis mais baixos de felicidade.
